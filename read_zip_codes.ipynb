```python
import pandas as pd
```

```python
dt = pd.DataFrame()
```

```python
xls = pd.ExcelFile('CPdescarga.xls')

states = xls.sheet_names[1:]
```

```python
states
```

```python
for (i, page) in enumerate(states):
    state = pd.read_excel('CPdescarga.xls', sheet_name=page)
    dt = pd.concat([dt, state])
```

```python
dt.fillna('', inplace=True)
```
